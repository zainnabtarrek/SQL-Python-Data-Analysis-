```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import pandasql as ps
import sklearn
```

```python
url="https://raw.githubusercontent.com/intern2grow/social-media-data-analysis/main/social_media_data.csv"
data=pd.read_csv(url)
data.head()
```

What was the total number of likes, shares, comments, and views (for videos) over the year?

```python
query="""
SELECT
     post_type,SUM(likes + shares + comments + views) AS total_posts
FROM
    data
where
    post_type = 'video'

"""
totalrevenue=ps.sqldf(query, locals())
print(totalrevenue)
```

Which platform had the highest engagement? (engagement = likes + shares + comments)

```python
query="""
SELECT
    platform,
    SUM(likes + shares + comments) AS engagement
FROM
    data
GROUP BY
    platform
ORDER BY
    engagement DESC
LIMIT 1;


"""
totalrevenue=ps.sqldf(query, locals())
print(totalrevenue)
```

What was the average number of likes, shares, comments, and views per post on each platform?

```python
query="""
SELECT
    platform,
   avg(likes),avg(likes),avg(shares),avg(comments),avg(views)
FROM
    data
GROUP BY
    platform
"""
totalrevenue=ps.sqldf(query, locals())
print(totalrevenue)
```

What was the total number of each type of post (video, image, text) on each platform?

```python
data.head()
```

```python
query = """
SELECT
    platform,
    post_type,
    COUNT(post_type) AS total
FROM
    data
WHERE
    platform != 'NonExistingPlatform'
GROUP BY
    platform,
    post_type
"""

totalrevenue = ps.sqldf(query, locals())
print(totalrevenue)
```

Which type of post had the highest average engagement?


```python
query = """
SELECT
    post_type,
    AVG(likes + comments + shares) AS avg_engagement
FROM
    data
GROUP BY
    post_type
ORDER BY
    avg_engagement DESC
LIMIT 1;
"""

totalrevenue = ps.sqldf(query, locals())
print(totalrevenue)
```

What was the total engagement in each quarter of the year?

```python
query="""
SELECT

  CASE
    WHEN strftime('%m', date) IN ('01', '02', '03') THEN 'Q1'
    WHEN strftime('%m', date) IN ('04', '05', '06') THEN 'Q2'
    WHEN strftime('%m', date) IN ('07', '08', '09') THEN 'Q3'
    WHEN strftime('%m', date) IN ('10', '11', '12') THEN 'Q4'
  END AS quarter,

    SUM(likes + shares + comments) AS total_engagement
FROM
    data
GROUP BY
    quarter
"""

totalrevenue = ps.sqldf(query, locals())
print(totalrevenue)
```
